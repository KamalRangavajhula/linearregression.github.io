```python
# dividing the label attribute i.e. the attribute which is dependent and the independent attributes
y = players['Rating']
X = players[['Acceleration', 'Speed','Stamina','Strength','Balance','Agility','Jumping','Heading','Shot_Power','Finishing']]
```

```python
sns.pairplot(X)
# the seaborn plot shows the relationship between the independent attributes which are used to predict the players ratings
```

```python
# splitting the dataset into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression
# importing the linear regression package from sckit learn 
```

```python
lm = LinearRegression()
# Creating an instance of a Linear regression
```

```python
lm.fit(X_train, y_train)
# Fitting the linear regression for the training data
```

```python
print('The coefficient values', lm.coef_)
# The linear regression coefficient values
```

```python
predictions = lm.predict(X_test)
# predicting the results of the testing data
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('Y test values')
plt.ylabel('Predicted linear regression values')
# Comparing the prediction vs the original values
```

```python
from sklearn import metrics
# The metrics which validates the performance of the regression model

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
